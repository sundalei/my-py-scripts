```python
"""
Elasticsearch document indexing notebook
"""

import json

import requests
from elastic_config import BASE_URL, CA_CERT_PATH, PASSWORD, USERNAME

%load_ext autoreload
%autoreload 2
```

## Indexing a new document into the `movies` index

```python
# pylint: disable=invalid-name

url = f"{BASE_URL}/movies/_doc/1"

payload: dict = {
    "title": "The Godfather",
    "synopsis": (
        "The aging patriarch of an organized crime dynasty "
        "transfers control of his clandestine empire to his reluctant son"
    ),
}

response = requests.put(
    url=url, auth=(USERNAME, PASSWORD), verify=CA_CERT_PATH, json=payload
)

print(json.dumps(response.json(), indent=2))
```

## Indexing a document with no ID using `POST`

## Fetching multiple documents at once

```python
# pylint: disable=invalid-name

url = f"{BASE_URL}/movies/_mget"

payload = {
    "ids": ["1", "12", "19", "34"],
}

response = requests.post(
    url=url, auth=(USERNAME, PASSWORD), verify=CA_CERT_PATH, json=payload
)

print(json.dumps(response.json(), indent=2))
```

## Fetching documents from three different indexes

```python
# pylint: disable=invalid-name

url = f"{BASE_URL}/_mget"

payload = {
    "docs": [
        {
            "_index": "tech_books",
            "_id": 3,
        },
        {
            "_index": "movies",
            "_id": 20,
        },
        {
            "_index": "flights",
            "_id": 2,
        },
    ],
}

response = requests.post(
    url=url, auth=(USERNAME, PASSWORD), verify=CA_CERT_PATH, json=payload
)

print(json.dumps(response.json(), indent=2))
```

## Using an `ids` query to fetch multiple documents

## Upsert block

```python
# pylint: disable=invalid-name

url = f"{BASE_URL}/movies/_update/5"

payload = {
    "script": {"source": "ctx._source.gross_earnings = '357.1m'"},
    "upsert": {"title": "Top Gun", "gross_earnings": "357.5m"},
}

response = requests.post(
    url=url, auth=(USERNAME, PASSWORD), verify=CA_CERT_PATH, json=payload
)

print(json.dumps(response.json(), indent=2))
```
