## Setup Environment

```python
"""
Elasticsearch indexing operation notebook
"""

import json

import requests
from elastic_config import BASE_URL, PASSWORD, USERNAME, VERIFY_CERT
```

## First document containing car data

```python
# pylint: disable=invalid-name

url = f"{BASE_URL}/cars/_doc/1"

payload: dict = {"make": "Maserati", "model": "GranTurismo Sport", "speed_mph": 186}

response = requests.put(
    url=url, auth=(USERNAME, PASSWORD), verify=VERIFY_CERT, json=payload
)

print(json.dumps(response.json(), indent=2))
```
